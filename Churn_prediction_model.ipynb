```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from imblearn.combine import SMOTEENN
```

```python
df=pd.read_csv('tel_churn.csv')
```

```python
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
X=df.drop('Churn',axis=1)
X
```

```python
y=df['Churn']
y
```

#### Train Test Split

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

#### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt= DecisionTreeClassifier(criterion='gini', random_state=100,min_samples_leaf = 8)
```

```python
model_dt.fit(X_train,y_train)
```

```python
y_pred_dt0= model_dt.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_dt0)
```

```python
print(classification_report(y_test,y_pred_dt0, labels=[0,1]))
```

```python
sm=SMOTEENN()
x_resampled,y_resampled=sm.fit_resample(X,y)
```

```python
X_train_sm,X_test_sm,y_train_sm,y_test_sm=train_test_split(x_resampled,y_resampled,test_size=0.2)
```

```python
model_dt_sm= DecisionTreeClassifier(criterion='gini', random_state=100,max_depth=6,min_samples_leaf = 8)
```

```python
model_dt_sm.fit(X_train_sm,y_train_sm)
```

```python
y_pred_dt_sm= model_dt.predict(X_test_sm)
```

```python
print(classification_report(y_test_sm,y_pred_dt_sm, labels=[0,1]))
```

```python
accuracy_score(y_test_sm,y_pred_dt_sm)
```

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators =100, criterion='gini', random_state=100,max_depth=6,min_samples_leaf=8)
```

```python
model_rf.fit(X_train_sm,y_train_sm)
```

```python
y_pred_rf= model_rf.predict(X_test_sm)
```

```python
print(classification_report(y_test_sm,y_pred_rf, labels=[0,1]))
```

```python
accuracy_score(y_test_sm,y_pred_rf)
```

#### Logistic Regeression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression(max_iter=500)
```

```python
model_lr.fit(X_train_sm,y_train_sm)
```

```python
y_pred_lr= model_lr.predict(X_test_sm)
```

```python
print(classification_report(y_test_sm,y_pred_lr, labels=[0,1]))
```

```python
accuracy_score(y_test_sm,y_pred_lr)
```

#### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
base_model_ab = DecisionTreeClassifier(max_depth=1)
```

```python
model_ab= AdaBoostClassifier( estimator= base_model_ab, n_estimators = 50, learning_rate=1,random_state=42) 
```

```python
model_ab.fit(X_train_sm,y_train_sm)
```

```python
y_pred_ab= model_ab.predict(X_test_sm)
```

```python
print(classification_report(y_test_sm,y_pred_ab, labels=[0,1]))
```

```python
accuracy_score(y_test_sm,y_pred_ab)
```
